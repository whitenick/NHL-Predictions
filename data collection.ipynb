```python
# games_df = pd.DataFrame(pd.read_json(json.dumps(dates[0]["games"])))
# gameFrames = []
games_df = pd.DataFrame()
for date in dates:
    games = date["games"]
    games_df = games_df.append(pd.read_json(json.dumps(games), orient='records'))

for date in dates2020:
    games = date["games"]
    games_df = games_df.append(pd.read_json(json.dumps(games), orient='records'))

len(games_df)
```

```python
def getAwayWins(df):
    return df['away']['leagueRecord']['wins']

def getAwayLosses(df):
    return df['away']['leagueRecord']['losses']

def getHomeWins(df):
    return df['home']['leagueRecord']['wins']

def getHomeLosses(df):
    return df['home']['leagueRecord']['losses']

def getTotalScore(df):
    return df['home']['score'] + df['away']['score']

def convertDate(df):
    ## 2018-10-03T23:00:00Z
    return datetime.strptime(df, "%Y-%m-%dT%H:%M:%SZ")

def getAwayTeamIds(df):
    return df['away']['team']['id']

def getHomeTeamIds(df):
    return df['home']['team']['id']

teams_df = games_df['teams']
games_df['Away_Team_ID'] = teams_df.apply(getAwayTeamIds)
games_df['Home_Team_ID'] = teams_df.apply(getHomeTeamIds)
games_df['Away_Team_Wins'] = teams_df.apply(getAwayWins)
games_df['Away_Team_Losses'] = teams_df.apply(getAwayLosses)
games_df['Home_Team_Wins'] = teams_df.apply(getHomeWins)
games_df['Home_Team_Losses'] = teams_df.apply(getHomeLosses)
games_df['gameDate'] = games_df['gameDate'].apply(convertDate)

games_df['Total_Goals'] = teams_df.apply(getTotalScore)

print(len(games_df))
games_df.to_csv('nhl_games_2019.csv')

games_df
```
